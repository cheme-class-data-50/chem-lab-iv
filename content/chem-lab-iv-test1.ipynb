```python
import numpy as np
from scipy.optimize import curve_fit
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import t

D = 0.0060  # m, tube diameter
rho = 1.2  # kg/m^3, air density
mu = 1.83e-5  # Pa s, air viscosity
rhow = 1000  # kg/m^3, water density
g = 9.8  # m/s^2, gravitational acceleration

#######################INPUT VALUES HERE###############################
v = np.array([9.82, 11.8, 13.8, 15.8, 17.7, 19.6])  # INPUT VELOCITY in m/s
dh = np.array([0.0198, 0.0283, 0.0376, 0.0479, 0.0585, 0.0695])  # INPUT HEIGHT DIFFERENCE in m
error_in_v = 5  # INPUT ERROR IN VELOCITY in %
error_in_dh = 5  # INPUT ERROR IN HEIGHT DIFFERENCE in %
#######################################################################

Re = v * D * rho / mu
P = rhow * g * dh
fexpt = P / (2 * rho * v**2 * (0.50 / D))

def calculate_error_bounds(v, dh):
    v_lower = v * (1-error_in_v/100)
    v_upper = v * (1+error_in_v/100)
    dh_lower = dh * (1-error_in_dh/100)
    dh_upper = dh * (1+error_in_dh/100)

    Re_lower = v_lower * D * rho / mu
    Re_upper = v_upper * D * rho / mu

    P_lower = rhow * g * dh_lower
    P_upper = rhow * g * dh_upper

    f_lower = P_lower / (2 * rho * v_upper**2 * (0.50 / D))
    f_upper = P_upper / (2 * rho * v_lower**2 * (0.50 / D))

    return Re_lower, Re_upper, f_lower, f_upper

Re_lower, Re_upper, f_lower, f_upper = calculate_error_bounds(v, dh)

f_theory = lambda Re: 0.0791 * Re**(-0.25)

def f_model(Re, C1, C2):
    return C1 * Re**C2

initial_guess = [0.08, -0.25]
params, covariance = curve_fit(f_model, Re, fexpt, p0=initial_guess)
C1, C2 = params
errors = np.sqrt(np.diag(covariance))
C1_error, C2_error = errors

# Monte Carlo simulation
def monte_carlo_analysis(v, dh, num_simulations=1000):
    np.random.seed(42)

    synthetic_C1 = []
    synthetic_C2 = []

    for _ in range(num_simulations):
        v_simulated = v * np.random.uniform(1-error_in_v/100, 1+error_in_v/100, size=len(v))
        dh_simulated = dh * np.random.uniform(1-error_in_dh/100, 1+error_in_dh/100, size=len(dh))

        Re_simulated = v_simulated * D * rho / mu
        P_simulated = rhow * g * dh_simulated
        f_simulated = P_simulated / (2 * rho * v_simulated**2 * (0.50 / D))

        try:
            params_simulated, _ = curve_fit(f_model, Re_simulated, f_simulated, p0=initial_guess)
            synthetic_C1.append(params_simulated[0])
            synthetic_C2.append(params_simulated[1])
        except:
            continue

    return np.array(synthetic_C1), np.array(synthetic_C2)

synthetic_C1, synthetic_C2 = monte_carlo_analysis(v, dh)

C1_lower = np.percentile(synthetic_C1, 2.5)
C1_upper = np.percentile(synthetic_C1, 97.5)
C2_lower = np.percentile(synthetic_C2, 2.5)
C2_upper = np.percentile(synthetic_C2, 97.5)

params_summary = pd.DataFrame({
    "Parameter": ["C1", "C2"],
    "Value": [round(C1, 4), round(C2, 3)],
    "95% CI Lower": [round(C1_lower, 3), round(C2_lower, 3)],
    "95% CI Upper": [round(C1_upper, 3), round(C2_upper, 3)]
})

plt.figure(figsize=(5, 5))
plt.errorbar(Re, fexpt, xerr=[Re - Re_lower, Re_upper - Re], yerr=[fexpt - f_lower, f_upper - fexpt], fmt='o', markersize=8, color='tab:blue', alpha=1, capsize=3)
plt.plot(Re, f_model(Re, C1, C2), color='tab:blue', linestyle='-', label='Best-fit model $f = C_1 \mathrm{Re}^{C_2}$')
plt.plot(Re, f_theory(Re), color='tab:orange', linestyle='--', label='$f_\mathrm{theory} = 0.0791 \mathrm{Re}^{-0.25}$')

# Customize the plot
plt.title('Analysis of friction factor for turbulent air flow in a plastic tube', pad=20)
plt.xlabel('Reynolds number (Re)')
plt.ylabel('Friction factor ($f$)')
plt.ylim(0.0, 0.02)
plt.xlim(0, 10000)
plt.yticks(np.arange(0.0, 0.021, 0.005))
plt.minorticks_on()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.grid(True, which='minor', linestyle=':', linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.savefig("friction_factor_plot.png", dpi=300)
plt.show()

plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
plt.hist(synthetic_C1, bins=30, color='tab:olive', alpha=0.7, edgecolor='black')
plt.axvline(C1, color='red', linestyle='--', label=f'Best-fit $C_1 = {C1:.4f}$')
plt.title('Distribution of $C_1$')
plt.xlabel('$C_1$')
plt.ylabel('Frequency')
plt.legend()

# Histogram for C2
plt.subplot(1, 2, 2)
plt.hist(synthetic_C2, bins=30, color='tab:cyan', alpha=0.7, edgecolor='black')
plt.axvline(C2, color='red', linestyle='--', label=f'Best-fit $C_2 = {C2:.3f}$')
plt.title('Distribution of $C_2$')
plt.xlabel('$C_2$')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()

# Display regression parameter summary
params_summary
```

```python
import numpy as np
from scipy.optimize import curve_fit
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import t

# Physical constants
D = 0.0060  # m, tube diameter
RHO_AIR = 1.2  # kg/m^3, air density
MU = 1.83e-5  # Pa s, air viscosity
RHO_WATER = 1000  # kg/m^3, water density
G = 9.8  # m/s^2, gravitational acceleration

# Input data
V = np.array([9.82, 11.8, 13.8, 15.8, 17.7, 19.6])  # velocity in m/s
DH = np.array([0.0198, 0.0283, 0.0376, 0.0479, 0.0585, 0.0695])  # height difference in m
ERROR_V_PERCENT = 5  # error in velocity (%)
ERROR_DH_PERCENT = 5  # error in height difference (%)

# Theoretical friction factor function
def f_theory(Re):
    return 0.0791 * Re**(-0.25)

# Model for curve fitting
def f_model(Re, C1, C2):
    return C1 * Re**C2

# Calculate Reynolds number, pressure, and experimental friction factor
def calculate_friction_parameters(v, dh):
    Re = v * D * RHO_AIR / MU
    P = RHO_WATER * G * dh
    f_expt = P / (2 * RHO_AIR * v**2 * (0.50 / D))
    return Re, P, f_expt

# Error propagation (corrected)
def calculate_error_bounds(v, dh, error_v_percent, error_dh_percent):
    v_lower = v * (1 - error_v_percent / 100)
    v_upper = v * (1 + error_v_percent / 100)
    dh_lower = dh * (1 - error_dh_percent / 100)
    dh_upper = dh * (1 + error_dh_percent / 100)

    # Corrected: Use v_lower for Re_lower and v_upper for Re_upper
    Re_lower, _, _ = calculate_friction_parameters(v_lower, dh_lower)
    Re_upper, _, _ = calculate_friction_parameters(v_upper, dh_upper)
    _, _, f_lower = calculate_friction_parameters(v_upper, dh_lower)  # Worst-case lower f
    _, _, f_upper = calculate_friction_parameters(v_lower, dh_upper)  # Worst-case upper f

    return Re_lower, Re_upper, f_lower, f_upper

# Monte Carlo simulation
def monte_carlo_analysis(v, dh, error_v_percent, error_dh_percent, num_simulations=1000):
    np.random.seed(42)
    synthetic_C1, synthetic_C2 = [], []
    initial_guess = [0.08, -0.25]

    for _ in range(num_simulations):
        v_sim = v * np.random.uniform(1 - error_v_percent/100, 1 + error_v_percent/100, size=len(v))
        dh_sim = dh * np.random.uniform(1 - error_dh_percent/100, 1 + error_dh_percent/100, size=len(dh))
        Re_sim, _, f_sim = calculate_friction_parameters(v_sim, dh_sim)

        try:
            params_sim, _ = curve_fit(f_model, Re_sim, f_sim, p0=initial_guess)
            synthetic_C1.append(params_sim[0])
            synthetic_C2.append(params_sim[1])
        except RuntimeError:
            continue

    return np.array(synthetic_C1), np.array(synthetic_C2)

# Main analysis
Re, P, f_expt = calculate_friction_parameters(V, DH)
Re_lower, Re_upper, f_lower, f_upper = calculate_error_bounds(V, DH, ERROR_V_PERCENT, ERROR_DH_PERCENT)

# Curve fitting
initial_guess = [0.08, -0.25]
params, covariance = curve_fit(f_model, Re, f_expt, p0=initial_guess)
C1, C2 = params
C1_error, C2_error = np.sqrt(np.diag(covariance))

# Monte Carlo simulation
synthetic_C1, synthetic_C2 = monte_carlo_analysis(V, DH, ERROR_V_PERCENT, ERROR_DH_PERCENT)
C1_ci = np.percentile(synthetic_C1, [2.5, 97.5])
C2_ci = np.percentile(synthetic_C2, [2.5, 97.5])

# Summary table
params_summary = pd.DataFrame({
    "Parameter": ["C1", "C2"],
    "Value": [f"{C1:.4f}", f"{C2:.3f}"],
    "95% CI Lower": [f"{C1_ci[0]:.3f}", f"{C2_ci[0]:.3f}"],
    "95% CI Upper": [f"{C1_ci[1]:.3f}", f"{C2_ci[1]:.3f}"]
})

# Plot 1: Friction Factor vs. Reynolds Number
plt.figure(figsize=(5, 5))
plt.errorbar(Re, f_expt,
             xerr=[Re - Re_lower, Re_upper - Re],
             yerr=[f_expt - f_lower, f_upper - f_expt],
             fmt='o', markersize=8, color='tab:blue', alpha=0.8, capsize=3, label='Experimental data')
plt.plot(Re, f_model(Re, C1, C2), color='tab:blue', lw=2, label=f'Fit: $f = {C1:.3f} \, \mathrm{{Re}}^{{{C2:.3f}}}$')
plt.plot(Re, f_theory(Re), color='tab:orange', ls='--', lw=2, label='$f_{theory} = 0.0791 \, \mathrm{Re}^{-0.25}$')

plt.title('Friction Factor vs. Reynolds Number', pad=15)
plt.xlabel('Reynolds Number (Re)')
plt.ylabel('Friction Factor ($f$)')
plt.xlim(0, 10000)
plt.ylim(0, 0.02)
plt.yticks(np.arange(0, 0.021, 0.005))
plt.grid(True, which='major', ls='--', lw=0.5)
plt.grid(True, which='minor', ls=':', lw=0.3)
plt.minorticks_on()
plt.legend()
plt.tight_layout()
plt.savefig("friction_factor_vs_reynolds.png", dpi=300, bbox_inches='tight')
plt.show()

# Plot 2: Monte Carlo Distribution of C1
plt.figure(figsize=(5, 5))
plt.hist(synthetic_C1, bins=30, color='tab:olive', alpha=0.7, edgecolor='black')
plt.axvline(C1, color='red', ls='--', label=f'$C_1 = {C1:.4f}$')
plt.title('Distribution of $C_1$')
plt.xlabel('$C_1$')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.savefig("distribution_C1.png", dpi=300, bbox_inches='tight')
plt.show()

# Plot 3: Monte Carlo Distribution of C2
plt.figure(figsize=(5, 5))
plt.hist(synthetic_C2, bins=30, color='tab:cyan', alpha=0.7, edgecolor='black')
plt.axvline(C2, color='red', ls='--', label=f'$C_2 = {C2:.3f}$')
plt.title('Distribution of $C_2$')
plt.xlabel('$C_2$')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.savefig("distribution_C2.png", dpi=300, bbox_inches='tight')
plt.show()

# Display results
print("\nRegression Parameter Summary:")
print(params_summary.to_string(index=False))
```
